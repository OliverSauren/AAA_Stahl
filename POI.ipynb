```python
import osmnx as ox
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Specify the name of the place
place_name = "Chicago, Illinois, USA"

# Download POIs of type 'restaurant'
tags = {'amenity': 'restaurant'}
gdf = ox.features_from_place(place_name, tags)

# Filter out only the Point geometries
gdf_points = gdf[gdf.geometry.type == 'Point']

# Convert the GeoDataFrame to a DataFrame and keep only the latitude and longitude
pois_df = pd.DataFrame({
    'latitude': gdf_points.geometry.y,
    'longitude': gdf_points.geometry.x
})

# Extract the start_latitude and start_longitude columns
taxi_locations = df_sample[['start_latitude', 'start_longitude']]

# Rename the columns to match the POIs DataFrame
taxi_locations.columns = ['latitude', 'longitude']

# Concatenate the taxi locations and the POIs
all_locations = pd.concat([taxi_locations, pois_df])

# Fit a KMeans model on the locations
kmeans = KMeans(n_clusters=5, random_state=0).fit(all_locations)

# Add the cluster labels to the dataframe
all_locations['cluster'] = kmeans.labels_

# Plot the clusters
plt.figure(figsize=(10, 10))
sns.scatterplot(x='longitude', y='latitude', hue='cluster', data=all_locations, palette='viridis')

# Set the limits of the plot to the area of interest
plt.xlim([-87.913624596, -87.529950466])
plt.ylim([41.660136051, 42.021223593])

plt.title('POI Clusters')
plt.show()
```

# Land-use clustering

```python
import osmnx as ox
from shapely.geometry import Point
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt


north, south, east, west = df_sample['end_latitude'].max(), df_sample['end_latitude'].min(), df_sample['end_longitude'].max(), df_sample['end_longitude'].min()

# Fetching the land-use polygons
land_use = ox.features_from_bbox(north, south, east, west, tags={'landuse': True})

# Function to calculate distance from a point to a land-use type's centroid
def distance_to_landuse(lat, lon, landuse_type):
    # Filter polygons of the given land-use type
    polygons = land_use[land_use['landuse'] == landuse_type]

    point = Point(lon, lat)

    if not polygons.empty:
        # Calculate distances to all centroids and return the minimum
        return point.distance(polygons.unary_union.centroid)
    else:
        # If there's no polygon of the given type, return a large distance (e.g., 9999)
        return 9999

# Apply the function for each land-use type
df_sample['distance_to_residential'] = df_sample.apply(lambda row: distance_to_landuse(row['end_latitude'], row['end_longitude'], 'residential'), axis=1)
df_sample['distance_to_commercial'] = df_sample.apply(lambda row: distance_to_landuse(row['end_latitude'], row['end_longitude'], 'commercial'), axis=1)
df_sample['distance_to_industrial'] = df_sample.apply(lambda row: distance_to_landuse(row['end_latitude'], row['end_longitude'], 'industrial'), axis=1)


# Features for clustering
X = df_sample[['distance_to_residential', 'distance_to_commercial', 'distance_to_industrial']]

# Scaling the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Applying K-Means clustering
kmeans = KMeans(n_clusters=5)  # Adjust the number of clusters as needed
df_sample['landuse_cluster'] = kmeans.fit_predict(X_scaled)

fig, ax = plt.subplots(figsize=(10,10))

colors = ['red', 'green', 'blue', 'yellow', 'purple']  # Adjust colors based on the number of clusters
for cluster_num in range(kmeans.n_clusters):
    cluster_data = df_sample[df_sample['landuse_cluster'] == cluster_num]
    ax.scatter(cluster_data['end_longitude'], cluster_data['end_latitude'], c=colors[cluster_num], label=f'Cluster {cluster_num}', s=10)

ax.set_title('Taxi Clusters based on Land-use Proximity')
ax.legend()
plt.show()
```

```python
from shapely.geometry import Point

# Function to calculate distance from a point to the closest polygon of a given land-use type
def distance_to_closest_landuse(lat, lon, landuse_type):
    # Filter polygons of the given land-use type
    polygons = land_use[land_use['landuse'] == landuse_type].geometry.values

    point = Point(lon, lat)

    if len(polygons) > 0:
        # Calculate distances to all polygons and return the minimum
        return min(point.distance(poly) for poly in polygons)
    else:
        # If there's no polygon of the given type, return a large distance (e.g., 9999)
        return 9999

# Apply the function for each land-use type
taxi_df['distance_to_residential'] = taxi_df.apply(lambda row: distance_to_closest_landuse(row['end_latitude'], row['end_longitude'], 'residential'), axis=1)
taxi_df['distance_to_commercial'] = taxi_df.apply(lambda row: distance_to_closest_landuse(row['end_latitude'], row['end_longitude'], 'commercial'), axis=1)
taxi_df['distance_to_industrial'] = taxi_df.apply(lambda row: distance_to_closest_landuse(row['end_latitude'], row['end_longitude'], 'industrial'), axis=1)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Features for clustering
X = taxi_df[['distance_to_residential', 'distance_to_commercial', 'distance_to_industrial']]

# Scaling the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Applying K-Means clustering
kmeans = KMeans(n_clusters=5)  # Adjust the number of clusters as needed
taxi_df['landuse_cluster'] = kmeans.fit_predict(X_scaled)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10,10))

colors = ['red', 'green', 'blue', 'yellow', 'purple']  # Adjust colors based on the number of clusters
for cluster_num in range(kmeans.n_clusters):
    cluster_data = taxi_df[taxi_df['landuse_cluster'] == cluster_num]
    ax.scatter(cluster_data['end_longitude'], cluster_data['end_latitude'], c=colors[cluster_num], label=f'Cluster {cluster_num}', s=10)

ax.set_title('Taxi Clusters based on Land-use Proximity')
ax.legend()
plt.show()
```
